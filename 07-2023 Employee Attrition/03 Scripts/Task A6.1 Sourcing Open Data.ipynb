### 03. Consistency Checks

```python
# Check for mixed column types
for col in df_ea.columns.tolist():
  weird = (df_ea[[col]].applymap(type) != df_ea[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_ea[weird]) > 0:
    print (col)
```
